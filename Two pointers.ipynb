Problem A. Merging arrays


You are given two arrays, sorted in non-decreasing order. Merge them into one sorted array.

Input
The first line contains integers n and m, the sizes of the arrays (1≤n,m≤10^5). The second line contains n integers ai, elements of the first array, the third line contains m integers bi, elements of the second array (−10^9≤ai,bi≤10^9).

Output
Print n+m integers, the merged array.

Example
Input
6 7
1 6 9 13 18 18
2 3 8 13 15 21 25

Output
1 2 3 6 8 9 13 13 15 18 18 21 25



```python
m,n = map(int, input().split())
a = list(map(int, input().split()))
b = list(map(int, input().split()))
i,j = 0,0
c = []
while i < len(a) and j < len(b):
    if a[i] < b[j]:
        c.append(a[i])
        i += 1
    else:
        c.append(b[j])
        j += 1
c.extend(a[i:])
c.extend(b[j:])
print(*c)
```

Problem C. Number of Equal

You are given two arrays aa and bb, sorted in non-decreasing order. Find the number of pairs (i,j) for which ai=bj.

Input
The first line contains integers nn and mm, the sizes of the arrays (1≤n,m≤10^5). The second line contains n integers ai, elements of the first array, the third line contains m integers bi, elements of the second array (−10^9≤ai,bi≤10^9).

Output
Print one number, the answer to the problem.

Example

Input
8 7

1 1 3 3 3 5 8 8

1 3 3 4 5 5 5

Output
11



```python
m,n = map(int, input().split())
a = list(map(int, input().split()))
b = list(map(int, input().split()))
d1 = {}
d2 ={}
for i in a:
    if i in d1:
        d1[i] += 1
    else:
        d1[i] = 1
for i in b:
    if i in d2:
        d2[i] += 1
    else:
        d2[i] = 1
ans = 0
for k in d1:
    if k in d2:
        ans += d1[k] * d2[k]
print(ans)
```

Problem D. Segment With Small Sum

Given an array of n integers ai. Let's say that the segment of this array a[l..r] (1≤l≤r≤n) is good if the sum of elements on this segment is at most s. Your task is to find the longest good segment.

Input    The first line contains integers nn and s (1≤n≤105, 1≤s≤1018). The second line contains integers ai (1≤ai≤109).

Output
Print one integer, the length of the longest good segment. If there are no such segments, print -1.

Example

Input

7 20

2 6 4 3 6 8 9

Output
4



```python
n,m = map(int, input().split())
a = list(map(int, input().split()))
l,ans = 0, 0
sum = 0
for r in range(n):
    sum += a[r]
    while sum > m:
        sum -= a[l]
        l += 1
    ans = max(ans,r - l + 1)

print(ans)


```

Problem E. Segment With big Sum

Given an array of n integers aiai. Let's say that the segment of this array a[l..r] (1≤l≤r≤n) is good if the sum of elements on this segment is at least s. Your task is to find the shortest good segment.

Input
The first line contains integers n and s (1≤n≤105, 1≤s≤1018). The second line contains integers ai (1≤ai≤109).

Output
Print one integer, the length of the shortest good segment. If there are no such segments, print −1.

Example

Input

7 20

2 6 4 3 6 8 9

Output   3



```python
n,m = map(int,input().split())
a = list(map(int, input().split()))
l,ans = 0, n + 1
sum = 0
for r in range(n):
    sum += a[r]
    while sum >= m:
        ans = min(ans,r - l + 1)
        sum -= a[l]
        l += 1
print(ans)

```
